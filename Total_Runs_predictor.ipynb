```python
import pandas as pd

# Create a new column 'Total_runs_scored' and initialize it with 0
df['Total_runs_scored'] = 0

# Group the dataframe by 'match_id' and 'innings'
grouped_data = df.groupby(['match_id', 'innings'])

# Iterate over each group
for name, group in grouped_data:
    # Calculate total runs scored for this group
    total_runs = group.loc[group['result'] != -1, 'result'].sum()
    # Update the 'Total_runs_scored' column for this group
    df.loc[(df['match_id'] == name[0]) & (df['innings'] == name[1]), 'Total_runs_scored'] = total_runs

# Now cricket_data dataframe will have the 'Total_runs_scored' for each innings of each match_id
```

```python
# Converting categorical features using OneHotEncoding method
dfn = pd.get_dummies(data=df, columns=['batting_team', 'bowling_team'])
dfn.columns
```

```python
import pandas as pd

# Assuming your dataframe is named 'cricket_data'

# Create a new column 'Cumulative_total_runs' and initialize it with 0
dfn['Cumulative_total_runs'] = 0

# Group the dataframe by 'match_id' and 'innings'
grouped_data = dfn.groupby(['match_id', 'innings'])

# Iterate over each group
for name, group in grouped_data:
    # Initialize cumulative total runs for this group
    cumulative_total_runs = 0
    # Iterate over each row in the group
    for index, row in group.iterrows():
        # If the result is not -1 (indicating not out), add the runs to cumulative total runs
        if row['result'] != -1:
            cumulative_total_runs += row['result']
        # Update the 'Cumulative_total_runs' column for this row
        dfn.at[index, 'Cumulative_total_runs'] = cumulative_total_runs

# Now cricket_data dataframe will have the 'Cumulative_total_runs' for each innings of each match_id
```

```python
import pandas as pd

# Assuming your dataframe is named 'cricket_data'

# Create a new column 'Wickets_fallen' and initialize it with 0
dfn['Wickets_fallen'] = 0

# Group the dataframe by 'match_id' and 'innings'
grouped_data = dfn.groupby(['match_id', 'innings'])

# Iterate over each group
for name, group in grouped_data:
    # Initialize wickets fallen for this group
    wickets_fallen = 0
    # Iterate over each row in the group
    for index, row in group.iterrows():
        # If the result is -1 (indicating out), increment the wickets_fallen count
        if row['result'] == -1:
            wickets_fallen += 1
        # Update the 'Wickets_fallen' column for this row
        dfn.at[index, 'Wickets_fallen'] = wickets_fallen

# Now cricket_data dataframe will have the 'Wickets_fallen' for each innings of each match_id
```

```python
# Linear Regression Model
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(X_train,y_train)
     
```

```python
# Predicting results
y_pred_lr = linear_regressor.predict(X_test)
```

```python
# Linear Regression - Model Evaluation
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse, accuracy_score
print("---- Linear Regression - Model Evaluation ----")
print("Mean Absolute Error (MAE): {}".format(mae(y_test, y_pred_lr)))
print("Mean Squared Error (MSE): {}".format(mse(y_test, y_pred_lr)))
print("Root Mean Squared Error (RMSE): {}".format(np.sqrt(mse(y_test, y_pred_lr))))
```

```python

# Decision Tree Regression Model
from sklearn.tree import DecisionTreeRegressor
decision_regressor = DecisionTreeRegressor()
decision_regressor.fit(X_train,y_train)
```

```python
# Predicting results
y_pred_dt = decision_regressor.predict(X_test)
```

```python

# Decision Tree Regression - Model Evaluation
print("---- Decision Tree Regression - Model Evaluation ----")
print("Mean Absolute Error (MAE): {}".format(mae(y_test, y_pred_dt)))
print("Mean Squared Error (MSE): {}".format(mse(y_test, y_pred_dt)))
print("Root Mean Squared Error (RMSE): {}".format(np.sqrt(mse(y_test, y_pred_dt))))
```

```python
# Random Forest Regression Model
from sklearn.ensemble import RandomForestRegressor
random_regressor = RandomForestRegressor()
random_regressor.fit(X_train,y_train)
```

```python
# Predicting results
y_pred_rf = random_regressor.predict(X_test)
```

```python
# Random Forest Regression - Model Evaluation
print("---- Random Forest Regression - Model Evaluation ----")
print("Mean Absolute Error (MAE): {}".format(mae(y_test, y_pred_rf)))
print("Mean Squared Error (MSE): {}".format(mse(y_test, y_pred_rf)))
print("Root Mean Squared Error (RMSE): {}".format(np.sqrt(mse(y_test, y_pred_rf))))
```

```python
# AdaBoost Model using Linear Regression as the base learner
from sklearn.ensemble import AdaBoostRegressor
adb_regressor = AdaBoostRegressor(base_estimator=linear_regressor, n_estimators=100)
adb_regressor.fit(X_train, y_train)
```

```python
# Predicting results
y_pred_adb = adb_regressor.predict(X_test)
```
