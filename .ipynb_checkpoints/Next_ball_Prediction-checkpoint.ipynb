```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('match_data.csv' , low_memory=False)
```

```python
df
```

```python
df.columns
```

```python
df['total_runs'] = df.apply(lambda x: x['runs_off_bat'] + x['extras'], axis=1)
df['isOut'] = df['player_dismissed'].apply(lambda x: 1 if type(x) == type('str') else 0)
```

```python
important_features = ['venue' , 'innings' , 'ball' , 'batting_team', 'bowling_team', 'striker', 'bowler',
       'total_runs' , 'isOut']

for features in df.columns :
    if (features in important_features):
        print(features  , " = Yes")
    else :
        print(features  , " = No")
        df = df.drop(features , axis = 1)
```

```python
df.head(10)
```

```python
df['venue'].value_counts()
```

```python
batsman_points=df.groupby('striker')['total_runs'].mean().to_dict()
batsman_points
```

```python
sorted_batsman_points = dict(sorted(batsman_points.items(), key=lambda item: item[1]))
```

```python
sorted_batsman_points
```

```python
df['batsman_point']=df['striker'].map(sorted_batsman_points)
```

```python
df.head()
```

```python
bowler_point=df.groupby('bowler')['isOut'].mean().to_dict()
sorted_bowler_point = dict(sorted(bowler_point.items(), key=lambda item: item[1]))
sorted_bowler_point
```

```python
df.head()
```

```python
df['bowler_point']=df['bowler'].map(sorted_bowler_point)
```

```python
df.head()
```

```python
df[df['isOut'] == 1]
```

```python
import numpy as np
df['result'] = np.where(df['isOut'] == 1, -1, df['total_runs'])
df.head()
```

```python
df[df['result'] == -1]
# Ignored the runs on the ball in which the player got out
```

```python
df['result'].unique()
```

```python
venue_point=df.groupby('venue')['result'].mean().to_dict()
sorted_venue_point = dict(sorted(venue_point.items(), key=lambda item: item[1]))
df['venue_point']=df['venue'].map(sorted_venue_point)
sorted_venue_point
```

```python
df.head()
```

```python
df['batting_team'].unique()
```

```python
consistent_teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
                    'Mumbai Indians', 'Kings XI Punjab', 'Royal Challengers Bangalore',
                    'Delhi Capitals', 'Sunrisers Hyderabad' ,'Gujarat Titans' ,'Lucknow Super Giants']
```

```python
print("Before_removing_inconsistent_teams : {}".format(df.shape))
df = df[(df['batting_team'].isin(consistent_teams)) & (df['bowling_team'].isin(consistent_teams))]
print("After_removing_inconsistent_teams : {}".format(df.shape))
```

```python
df.head()
```

```python
new_important_features = ['innings' , 'ball' , 'batting_team', 'bowling_team',
       'batsman_point'	,'bowler_point'	,'result'	,'venue_point']

for features in df.columns :
    if (features in new_important_features):
        print(features  , " = Yes")
    else :
        print(features  , " = No")
        df = df.drop(features , axis = 1)
```

```python
df.head()
```

```python
# Selecting correlated features using Heatmap
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns
numerical_df = df.select_dtypes(include='number')

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
# Converting categorical features using OneHotEncoding method
dfn = pd.get_dummies(data=df, columns=['batting_team', 'bowling_team'])
dfn.columns
```

```python
dfn.shape
```

```python
dfn = dfn[['innings', 'ball', 'batsman_point', 'bowler_point',
       'venue_point', 'batting_team_Chennai Super Kings',
       'batting_team_Delhi Capitals', 'batting_team_Gujarat Titans',
       'batting_team_Kings XI Punjab', 'batting_team_Kolkata Knight Riders',
       'batting_team_Lucknow Super Giants', 'batting_team_Mumbai Indians',
       'batting_team_Rajasthan Royals',
       'batting_team_Royal Challengers Bangalore',
       'batting_team_Sunrisers Hyderabad', 'bowling_team_Chennai Super Kings',
       'bowling_team_Delhi Capitals', 'bowling_team_Gujarat Titans',
       'bowling_team_Kings XI Punjab', 'bowling_team_Kolkata Knight Riders',
       'bowling_team_Lucknow Super Giants', 'bowling_team_Mumbai Indians',
       'bowling_team_Rajasthan Royals',
       'bowling_team_Royal Challengers Bangalore',
       'bowling_team_Sunrisers Hyderabad','result']]
```

```python
for column in dfn.columns :
    if(dfn[column].dtype == bool) :
        dfn[column]= dfn[column].astype(int)
```

<h2>Data Spliting</h2>

```python
X = dfn.iloc[: , :-1]
y = dfn.iloc[: , -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.20 , random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

<h2>Model Training</h2>

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression


models = {
    'Random Forest' : RandomForestClassifier(),
    'Decision Tree' : DecisionTreeClassifier(),
    'Logistic Regression' : LogisticRegression()
}


from sklearn.metrics import accuracy_score
```

```python
model = list(models.values())
model_names = list(models.keys())
print(model)
print(model_names)
```

```python
for  i in range(len(models)) :
    model[i].fit(X_train , y_train)
    y_pred = model[i].predict(X_test)
    score = accuracy_score(y_test , y_pred)
    print(f'score of model {model_names[i]} = ', score)
    print('\n')
    
```

```python
classfier=RandomForestClassifier()
```

```python
## Hypeparameter Tuning
params={'max_depth':[3,5,10,None],
              'n_estimators':[100,200,300],
               'criterion':['gini','entropy']
              }
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
cv=RandomizedSearchCV(classfier,param_distributions=params,scoring='accuracy',cv=5,verbose=3)
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
Rnf = RandomForestClassifier(max_depth= 10 , n_estimators= 100 , criterion= 'gini')
```

```python
Rnf
```

```python
final_model = Rnf.fit(X_train , y_train)
final_model
```

```python
final_y_pred=final_model.predict(X_test)
final_score= accuracy_score(y_test  , final_y_pred)
final_score
```
